# Time peiod and PVM Analysis

```python
#time period comparison setup
path_date = r'C:\Users\Lenovo\Documents\GitHub\FMCG-Sales-PVM-Analytics\data\schema\dim_date.csv'
df_date = pd.read_csv(path_date,sep=';',quotechar='"')
df_date
```

```python
years = sorted(df_date['yr'].unique())
years
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
pvm_result  = []

for yr in years[1:]:
    base_yr = df[df['order_date'].dt.year==yr-1]
    current_yr = df[df['order_date'].dt.year==yr]

    #Calculation at SKU level
    base_agg = base_yr.groupby('product_id').agg(base_qty=('quantity','sum'),base_sales=('sales','sum'),base_cogs=('total_COGS','sum')).reset_index()
    curr_agg = current_yr.groupby('product_id').agg(curr_qty=('quantity','sum'),curr_sales=('sales','sum'),curr_cogs=('total_COGS','sum')).reset_index()

    pvm = pd.merge(base_agg,curr_agg,on='product_id',how='outer').fillna(0)

    pvm['base_price'] = pvm['base_sales']/pvm['base_qty']
    pvm['curr_price'] = pvm['curr_sales']/pvm['curr_qty']
    pvm = pvm.replace([float('inf'),-float('inf')],0).fillna(0)

    #Impacts
    price_impact = (pvm['base_qty']*(pvm['curr_price']-pvm['base_price'])).sum()
    vol_impact = (pvm['base_price']*(pvm['curr_qty']-pvm['base_qty'])).sum()
    mix = ((pvm['base_qty']-pvm['curr_qty'])*(pvm['base_price']-pvm['curr_price'])).sum()
    cost = -(pvm['base_cogs']-pvm['curr_cogs']).sum()


    pvm_result.append({
        'year' : yr,
        'base_margin' : (pvm['base_sales'] - pvm['base_cogs']).sum(),
        'price_impact' : price_impact,
        'volume_impact' : vol_impact,
        'mix' : mix,
        'cost' : cost,
        'total_margin_change' : price_impact+vol_impact+mix+cost,
        'current_margin' : (pvm['curr_sales'] - pvm['curr_cogs']).sum()
    })
```

```python
pvm_summary = pd.DataFrame(pvm_result)
pvm_summary
```

```python
pvm_summary_transposed = pvm_summary.transpose()
pvm_summary_transposed
```

```python
pvm_summary_transposed.reset_index(inplace=True)
pvm_summary_transposed.rename(columns={'index':'Step'},inplace=True)
```

```python
df_melted = pvm_summary_transposed.melt(id_vars='Step',var_name='Year',value_name='Value')
df_melted
```

```python
step_map = {
    'base_margin' : 'Base Margin',
    'price_impact' : 'Price Impact',
    'volume_impact' : 'Volume Impact',
    'mix' : 'Mix Impact',
    'cost' : 'Cost Impact',
    'total_margin_change' : 'Total Margin Change',
    'current_margin' : 'Current Margin'
}

df_melted['Step'] = df_melted['Step'].map(step_map)
```

```python
order_map = {
    'Base Margin': 1,
    'Price Impact': 2,
    'Volume Impact': 3,
    'Mix Impact': 4,
    'Cost Impact': 5,
    'Total Margin Change' : 6,
    'Current Margin': 7
}

df_melted['Order'] = df_melted['Step'].map(order_map)
```

```python
df_melted = df_melted[~df_melted['Step'].isnull()]
```
