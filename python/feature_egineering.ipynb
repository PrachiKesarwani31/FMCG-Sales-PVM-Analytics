# Time peiod and PVM Analysis

```python
#time period comparison setup
path_date = r'C:\Users\Lenovo\Documents\GitHub\FMCG-Sales-PVM-Analytics\data\schema\dim_date.csv'
df_date = pd.read_csv(path_date,sep=';',quotechar='"')
df_date
```

```python
years = sorted(df_date['yr'].unique())
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
for yr in years[1:]:
    base_yr = df[df['order_date'].dt.year==yr-1]
    current_yr = df[df['order_date'].dt.year==yr]

    #Calculation at SKU level
    base_agg = base_yr.groupby('product_id').agg(base_qty=('quantity','sum'),base_sales=('sales','sum'),base_cogs=('total_COGS','sum')).reset_index()
    curr_agg = current_yr.groupby('product_id').agg(curr_qty=('quantity','sum'),curr_sales=('sales','sum'),curr_cogs=('total_COGS','sum')).reset_index()

    pvm = pd.merge(base_agg,curr_agg,on='product_id',how='outer').fillna(0)

    pvm['base_priice'] = pvm['base_sales']/pvm['base_qty']
    pvm['curr_price'] = pvm['curr_sales']/pvm['curr_qty']
    pvm = pvm.replace([float('inf'),-float('inf')],0).fillna(0)

    #Impacts
    
```
